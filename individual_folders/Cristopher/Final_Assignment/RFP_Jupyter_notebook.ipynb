##### Heat Island data with Open Weather Map API

```bash
import matplotlib.pyplot as plt
import pandas as pd
import requests

# Define the URL for the heat island data
url = "https://api.openweathermap.org/data/2.5/weather"
api_key = "d15258c57b8467e9abbd7ce20ed95c35"

# Define the cities array
cities = ("London,UK" "Rome,IT" "Tokyo,JP", "Paris,FR", "Sydney,AU", "New York,US", "Los Angeles,US", "Beijing,CN", "Moscow,RU", "Cairo,EG", "Rio de Janeiro,BR", "Cape Town,ZA")

# Make the API request and display the weather data
response = requests.get(f"{url}?appid={api_key}&q={cities[0]}")
weather_data = response.json()

#Store the temperature and humidity data
temperature = weather_data['main']['temp']
humidity = weather_data['main']['humidity']

# Define the cities array
cities = ("London,UK", "Rome,IT", "Tokyo,JP", "Paris,FR", "Sydney,AU", "New York,US", "Los Angeles,US", "Beijing,CN", "Moscow,RU", "Cairo,EG", "Rio de Janeiro,BR", "Cape Town,ZA")

# Define empty lists to store temperature and humidity data
temperatures = []
humidities = []

# Iterate over the cities array and make API requests for each city
for city in cities:
    response = requests.get(f"{url}?appid={api_key}&q={city}")
    weather_data = response.json()
    
    # Extract temperature and humidity data from the response
    temperature = weather_data['main']['temp']
    humidity = weather_data['main']['humidity']
    
    # Append the data to the respective lists
    temperatures.append(temperature)
    humidities.append(humidity)

# Plot the temperature and humidity data
plt.figure(figsize=(10, 6))
plt.bar(cities, temperatures, label='Temperature (K)')
plt.bar(cities, humidities, label='Humidity (%)')

# Add labels and title to the plot
plt.xlabel('Cities')
plt.ylabel('Temperature (K) / Humidity (%)')
plt.title('Temperature and Humidity by Cities')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

### Use Open Weather Map API
##### Use of multiple cities

```bash
#Bash

#!/bin/bash

# Define the API endpoint and parameters
url="https://api.openweathermap.org/data/2.5/weather"
api_key="d15258c57b8467e9abbd7ce20ed95c35"
cities=("London,UK" "Rome,IT" "Tokyo,JP", "Paris,FR", "Sydney,AU", "New York,US", "Los Angeles,US", "Beijing,CN", "Moscow,RU", "Cairo,EG", "Rio de Janeiro,BR", "Cape Town,ZA")

# Iterate over the cities and fetch weather data
for city in "${cities[@]}"; do
    params="q=${city}&appid=${api_key}&units=metric"

    # Send a GET request to the API endpoint
    response=$(curl -s "${url}?${params}")

    # Check if the request was successful
    if [[ $? -eq 0 ]]; then
        # Parse the JSON response
        temperature=$(echo "${response}" | jq -r '.main.temp')
        humidity=$(echo "${response}" | jq -r '.main.humidity')
        description=$(echo "${response}" | jq -r '.weather[0].description')

        # Print the weather data for the city
        echo "Weather in ${city}:"
        echo "Temperature: ${temperature}°C"
        echo "Humidity: ${humidity}%"
        echo "Description: ${description}"
        echo "------------------------"
    else
        echo "Failed to fetch weather data for ${city}"
    fi
done
```
